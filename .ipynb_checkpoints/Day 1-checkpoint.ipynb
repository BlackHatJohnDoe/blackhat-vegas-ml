In probability terms..

If
* A = Website is phishing
* B = Website has @ symbol

and

$$P(A|B) = 0.56797$$

$$P(B|A) = 0.86714$$

Let's do a quick sanity check and see if Bayes Formula holds here.
For those who are unfamiliar or had too fun last night at the slots..

### Bayes Formula
$$P(A|B) = \frac{P(B|A)*P(A)}{P(B)}$$

```python
p_b_a = num_phishing_and_at_symbol/float(num_phishing)  # P(B|A)
p_a_b = num_phishing_and_at_symbol/float(num_at_symbol) # P(A|B)
p_a = num_phishing / float(phishing.shape[0])           # P(A)
p_b = num_at_symbol / float(phishing.shape[0])          # P(B)
```

```python
p_a_b - (p_b_a*p_a)/p_b  # which evaluates to 1.11*10^-16, which is basically 0, up to floating point division
# Yay math still works, good to know
```

### Primer on Anomoly Detection

“An outlier is an observation in a data set which appears to be inconsistent with the remainder of that set of data.” 
-- Johnson 1992

#### Supervised Anomoly Detection
* Train from both normal data and anomolies
* Basically a standard Machine Learning problem

#### Semi-supervised Anomoly Detection (Novelty Detection)
* Only normal data given to train from
* Algorithms must use normal data to predict what is "not normal"

#### Unsupervised Anomoly Detection (Outlier Detection)
* No labels given whatsoever on our training set. Algorithm must decide what is normal AND an anomoly
* Algorithms make the asumption that anomalies are rare and are substantially different from the normal data

# Linear Models

Linear models are learning algorithms that assume that the target value is a linear combination of the input variables. In mathematical notion, if $\hat{y} is the predicted value, then

$$\hat{y}(w, x) = w_0 + w_1 x_1 + ... + w_p x_p$$

Across the module, we designate the vector $w = (w_1,..., w_p)$ as coef_ and $w_0$ as intercept_.

Today we will look at using linear models to perform both classification and regression.




```python
# http://scikit-learn.org/stable/modules/linear_model.html 
Linear Models
```
